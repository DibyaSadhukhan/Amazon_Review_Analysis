```python
#function to get the reveiw link of the product who have atleast 1 reveiw
def get_product_reveiw_link(startlink):
  headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36'}
  webPage = requests.get('/'.join(startlink.split('/')[:-1]), headers = headers).text
  soup = BeautifulSoup(webPage, 'lxml')
  tags=soup.find('span', class_='cr-widget-FocalReviews')
  table=soup.find('table', class_='a-keyvalue prodDetTable')
  time.sleep(random.randint(1,3))
  try:
    rev_link="https://www.amazon.in"+tags.find('a',class_="a-link-emphasis a-text-bold")['href']
  except:
    rev_link=""
  try:
    sent=""
    for tr in table.find_all('tr'):
      th=tr.find('th')
      td =tr.find('td')
      sent+=th.text.strip()+':'+td.text.strip()+'||'
    import os.path
    from os import path
    dest_folder = '/content/gdrive/MyDrive/Amazon_product_Details'
    if path.exists(dest_folder) == False:
      os.mkdir(dest_folder)
    with open('/content/gdrive/MyDrive/Amazon_product_Details/'+startlink.split('/')[-2]+'.txt', 'w+') as f:    
          f.write(sent[:-2])
    f.close()
    return rev_link,('/content/gdrive/MyDrive/Amazon_product_Details/'+startlink.split('/')[-2]+'.txt')
  except:
    return rev_link,""
```

```python
#function to scrape the reveiws and all relevant details from the reveiw 
def get_reveiws(startLink):
  reveiws=[]
  count=1
  headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.107 Safari/537.36'}
  webPage = requests.get(startLink, headers = headers).text
  soup = BeautifulSoup(webPage, 'lxml')
  #iterating through all the products 
  while soup.find('li',class_='a-last').find('a',href=True) != None:
  #while soup.find('li',class_='a-last')!= None:
    next=soup.find('li',class_='a-last').find('a',href=True)['href']
    next='https://www.amazon.in'+next
    for tags in soup.find_all('div',class_ = 'a-row a-spacing-small review-data'):
      buyer=tags.find_previous('div').find_previous('div').find_previous('div').text.strip()
      product_brought=tags.find_previous('div').find_previous('div').text.strip()
      reveiw=str(tags.text.strip())
      reveiw=reveiw.replace('\n',' ')
      #print(reveiw)
      reveiws.append((count,buyer,str(product_brought.split('\n')[0]),str(product_brought.split('\n')[1]),reveiw))
      count+=1  
    time.sleep(random.randint(1,3))
    webPage = requests.get(next, headers = headers).text
    soup = BeautifulSoup(webPage, 'lxml')
  for tags in soup.find_all('div',class_ = 'a-row a-spacing-small review-data'):
    buyer=tags.find_previous('div').find_previous('div').find_previous('div').text.strip()
    product_brought=tags.find_previous('div').find_previous('div').text.strip()
    reveiw=str(tags.text.strip())
    reveiw=reveiw.replace('\n',' ')
    reveiws.append((count,buyer,str(product_brought.split('\n')[0]),str(product_brought.split('\n')[1]),reveiw))
    count+=1    
  print(next)
  #returning the links of all the products available for the given query link
  return reveiws
```

```python
#function to write the scraped reveiw in a file
def write_reveiws_to_file(product,reveiws):
  import os.path
  from os import path
  dest_folder = '/content/gdrive/MyDrive/Amazon_product_Reveiws'
  if path.exists(dest_folder) == False:
    os.mkdir(dest_folder)
  
  with open('/content/gdrive/MyDrive/Amazon_product_Reveiws/'+product+'.txt', 'w+') as f:
        
      # write elements of list
      for items in reveiws:
          f.write((str(items)+"||"))
  # close the file
  f.close()
  return '/content/gdrive/MyDrive/Amazon_product_Reveiws/'+product+'.txt'
  
```
